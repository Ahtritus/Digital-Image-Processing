```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
import math
```

```python
img=cv2.imread('lena512color.tiff',0)
height,width=img.shape
border_img=cv2.copyMakeBorder(img,1,1,1,1,cv2.BORDER_CONSTANT,value=0)
plt.figure(figsize=(8,8))
plt.imshow(border_img,'gray')
```

```python
blur_kernel=np.array([[1,4,1],[4,8,4],[1,4,1]])
blur_img=np.zeros(img.shape,'uint8')
```

```python
for i in range(height):
    for j in range(width):
        blur_img[i,j]=int(np.sum(border_img[i:i+3,j:j+3]*blur_kernel)/np.sum(blur_kernel))
plt.figure(figsize=(10,10))
plt.imshow(blur_img,'gray')
```

```python
sobel_border_img=cv2.copyMakeBorder(blur_img,1,1,1,1,cv2.BORDER_CONSTANT,value=0)
sobelx=np.array([[-1,0,1],[-4,0,4],[-1,0,1]])
sobely=np.array([[-1,-4,-1],[0,0,0],[1,4,1]])
sobel_edge=np.zeros(img.shape,'uint8')
```

```python
for i in range(height):
    for j in range(width):
        xsum=np.sum(sobel_border_img[i:i+3,j:j+3]*sobelx)
        ysum=np.sum(sobel_border_img[i:i+3,j:j+3]*sobely)
        sobel_edge[i,j]= round(math.sqrt(xsum*xsum+ysum*ysum))
        
plt.figure(figsize=(10,10))
plt.imshow(sobel_edge,'gray')
```
