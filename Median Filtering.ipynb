```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread('noise_image.png',0)
plt.figure(figsize=(8,8))
plt.imshow(img,'gray')
```

```python
height,width=img.shape
border_img=cv2.copyMakeBorder(img,2,2,2,2,cv2.BORDER_CONSTANT,value=0)
blur_img=np.zeros(img.shape,'uint8')
median_arr=np.zeros((7,7))
```

```python
for i in range(height):
    for j in range(width):
        median_arr=border_img[i:i+7,j:j+7]
        np.sort(median_arr)
        blur_img[i,j]=np.median(median_arr)
plt.figure(figsize=(8,8))
plt.imshow(blur_img,'gray')
```

```python
#using library function
final = cv2.medianBlur(img, 7)
plt.figure(figsize=(8,8))
plt.imshow(final,'gray')
```
